```python
import numpy as np 
import pandas as pd 
```

```python
# import files
master = pd.read_excel("current_master_11_17_20.xlsx")
matches = pd.read_excel("AIA on victoria_11_14_20_3.xlsx")
```

```python
# oct 2020

# enter in last used match and household numbers from the numeration file 
match_number = 2468
household_match_number = 135

for i in range(0, len(matches.index), 2):
    
    ### define some variables 
    # find first record in the master
    unique_id_1 = matches.loc[matches.index[i], 'unique_id']   
    master_index_1 = master[master['unique_id'] == unique_id_1].index.tolist()
    # pull out match number and household id
    match_1 = master.loc[master_index_1, 'Joint ID for Matched Records'].item()
    household_match_1 = master.loc[master_index_1, 'Household Joint ID'].item()
    
    # find second record in the master
    unique_id_2 = matches.loc[matches.index[i + 1], 'unique_id']   
    master_index_2 = master[master['unique_id'] == unique_id_2].index.tolist()
    # pull out match number and household id
    match_2 = master.loc[master_index_2, 'Joint ID for Matched Records'].item()
    household_match_2 = master.loc[master_index_2, 'Household Joint ID'].item()
    
    ### generate match ids 
    # if both have existing match numbers, this could be a problem or could be something that should be examined
    if isinstance(match_1, int) & isinstance(match_2, int):
        if match_1 != match_2:
            my_string = "examine " + str(match_1) + " & " + str(match_2)
            master.loc[master.index[master_index_1], 'Joint ID for Matched Records'] = my_string
    # if only match 1 has an existing match number
    elif isinstance(match_1, int):
        master.loc[master.index[master_index_2], 'Joint ID for Matched Records'] = match_1
    # if only match 2 has an existing match number
    elif isinstance(match_2, int):
        master.loc[master.index[master_index_1], 'Joint ID for Matched Records'] = match_2
    # if neither have an existing match number, generate a new one
    else:
        master.loc[master.index[master_index_1], 'Joint ID for Matched Records'] = match_number
        master.loc[master.index[master_index_2], 'Joint ID for Matched Records'] = match_number
        match_number = match_number + 1
       
    ### repeat process for household ids
    # if both have existing match numbers, this could be a problem or could be something that should be examined
    if isinstance(household_match_1, int) & isinstance(household_match_2, int):
        if household_match_1 != household_match_2:
            my_string = "examine " + str(household_match_1) + " & " + str(household_match_2)
            master.loc[master.index[master_index_1], 'Household Joint ID'] = my_string
    # if only match 1 has an existing match number
    elif isinstance(household_match_1, int):
        master.loc[master.index[master_index_2], 'Household Joint ID'] = household_match_1
    # if only match 2 has an existing match number
    elif isinstance(household_match_2, int):
        master.loc[master.index[master_index_1], 'Household Joint ID'] = household_match_2
```
