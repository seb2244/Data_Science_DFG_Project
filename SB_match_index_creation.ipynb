```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv("1860-1930 database for family tracking Adam manual AFAM CANADA current-2.csv")
```

```python
raw_data.head()
```

```python
match_data = pd.read_csv("August 2019 Windsor with matches confirmed map.csv")
match_data.head()
```

```python
# convert to string and get rid of decimal points
count = 0
for year in raw_data['Census.Year']:
    try:
        raw_data.loc[count, 'Census.Year'] = str(int(year)).strip()
    except:
        raw_data.loc[count, 'Census.Year'] = str(year).strip()
    count = count + 1
    
# convert to string and get rid of decimal points
count = 0
for year in raw_data['CalculatedBirthYear']:
    try:
        raw_data.loc[count, 'CalculatedBirthYear'] = str(int(year)).strip()
    except:
        raw_data.loc[count, 'CalculatedBirthYear'] = str(year).strip()
    count = count + 1
```

```python
# convert to string to avoid nan errors 
match_data['CalculatedBirthYear'] = [str(year).strip() for year in match_data['CalculatedBirthYear']]
match_data['Census.Year'] = [str(year).strip() for year in match_data['Census.Year']]
```

```python
# get rid of all leading and trailing spaces in names 

match_data['First.Name'] = [str(name).strip() for name in match_data['First.Name']]
match_data['Last.Name'] = [str(name).strip() for name in match_data['Last.Name']]
raw_data['First.Name'] = [str(name).strip() for name in raw_data['First.Name']]
raw_data['Last.Name'] = [str(name).strip() for name in raw_data['Last.Name']]
```

```python
# merge
new_data = raw_data.merge(match_data, how = "left", on = ["Last.Name", "First.Name", "Census.Year", "CalculatedBirthYear"])
```

```python
# drop unnecessary columns
new_data = new_data.drop(new_data.loc[:, "State/Province_y":"Unnamed: 25"].columns, axis = 1)
```
